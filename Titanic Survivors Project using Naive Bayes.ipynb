```python
import pandas as pd
titanic = pd.read_csv('C:\\Users\\Mwaniki\\Desktop\\data science\\titanic.csv.csv')
titanic.head()
```

```python
#cleaning data

titanic.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
titanic.head()
```

```python
## Separating into dependent and independent categorizes

target = titanic.Survived
tita = titanic.drop('Survived',axis='columns')
```

```python
 ## changing sex category into numerals
    
dummies = pd.get_dummies(tita.Sex) 
dummies.head()
```

```python
tita = pd.concat([tita,dummies],axis='columns')
tita.head()
```

```python
tita.drop('Sex',axis='columns',inplace=True)
tita.head()
```

```python
tita.columns[tita.isna().any()]
```

```python
tita.Age = tita.Age.fillna(tita.Age.mean())
tita.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tita,target, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
